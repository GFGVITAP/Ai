```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Read the dataset from a CSV file

data = pd.read_csv("DailyDelhiClimateTrain.csv")

# Display the first few rows of the dataset

print(data.head())
```

```python
# Generate summary statistics for numeric columns

print(data.describe())
```

```python
# Display information about the dataset, including data types and non-null counts

print(data.info())
```

```python
# Create a Plotly Express line plot for mean temperature over the years

figure = px.line(data, x="date",
                 y="meantemp",
                 title='Mean Temperature in Delhi Over the Years')
figure.show()
```

```python
# Create a Plotly Express line plot for humidity over the years

figure = px.line(data, x="date",
                 y="humidity",
                 title='Humidity in Delhi Over the Years')
figure.show()
```

```python
# Create a Plotly Express line plot for wind speed over the years

figure = px.line(data, x="date",
                 y="wind_speed",
                 title='Wind Speed in Delhi Over the Years')
figure.show()
```

```python
# Create a scatter plot to explore the relationship between temperature and humidity

figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp",
                    trendline="ols",
                    title = "Relationship Between Temperature and Humidity")
figure.show()
```

```python
# Convert the "date" column to a datetime format and extract year and month

data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

```python
# Set the style and create a Seaborn line plot for temperature change over the years

plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

```python
# Rename columns for use with the Prophet forecasting model

forecast_data = data.rename(columns = {"date": "ds",
                                       "meantemp": "y"})
print(forecast_data)
```

```python
# Perform time series forecasting using Facebook Prophet

from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
